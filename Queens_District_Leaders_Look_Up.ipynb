```python
import pandas as pd
```

```python
# load the voter registeration file
df = pd.read_csv('Queens_voters_geocodio.csv')
df_queens_county_leaders = pd.read_csv('Queens_County_District_Leaders.csv')
df_queens_county_leaders
df.head()
```

```python
df.columns
```

```python
df_selected = df.loc[:, ['surname', 'first', 'middle', 'address', 'telephone_optional', 'email_address']]
df_selected
```

```python
df_queens_county_leaders
```

```python
df_queens_county_leaders[['First_Name', 'Middle_Name']] = df_queens_county_leaders['First Name'].str.split(expand=True)
df_queens_county_leaders['Middle_Name'] = df_queens_county_leaders['Middle_Name'].str.replace('.', '')
df_queens_county_leaders['Middle_Name'] = df_queens_county_leaders['Middle_Name'].fillna('')
df_queens_county_leaders = df_queens_county_leaders.applymap(lambda x: x.strip() if isinstance(x, str) else x)
#df_queens_county_leaders[['First_Name', 'Last Name']] = df_queens_county_leaders[['First_Name', 'Last Name']].applymap(str.lower)
df_queens_county_leaders['ID'] = (df_queens_county_leaders['First_Name'] + df_queens_county_leaders['Middle_Name'] + df_queens_county_leaders['Last Name']).str.lower()
df_queens_county_leaders['ID'] = df_queens_county_leaders['ID'].str.strip()
df_queens_county_leaders
```

```python
df_selected
```

```python
df_selected.dtypes
```

```python
# remember to strip white space for all the string column. space is bad for joining later
df_selected = df_selected.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#df_selected = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
#df_selected[['surname','first']] = df_selected[['surname','first']].applymap(str.strip)
df_selected['middle'] = df_selected['middle'].fillna('')
df_selected['ID'] = (df_selected['first'] + df_selected['middle'] + df_selected['surname']).str.lower()
df_selected['ID'] = df_selected['ID'].fillna('')
df_selected['ID'] = df_selected['ID'].str.strip()
df_selected
```

```python
#df_selected['first'] = df_selected['first'].str.strip()
df_selected[(df_selected['surname']=='LIU') & (df_selected['first']=='JOHN')]
#sum(df_selected['ID'].isna())
df_selected[df_selected['ID']=='johnliu']
```

```python
# Perform a left join
merged_df = pd.merge(df_queens_county_leaders, df_selected, on='ID', how='left')
#merged_df[merged_df['ID']=='johnliu']
```

```python
filtered_df = merged_df[merged_df['surname'].notna()]
filtered_df[filtered_df['ID']=='mufazzalhossain']
filtered_df['ID'].value_counts()
```
